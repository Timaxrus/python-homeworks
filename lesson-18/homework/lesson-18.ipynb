```python
import pandas as pd

df = pd.read_csv('stackoverflow_qa.csv')
df.head()

# Check if 'creationdate' column is of proper datetime datatype
# print(df['creationdate'].dtype)

# Check if there are any NaN values in the column
# print(df['creationdate'].isnull().sum())

# Convert the object datatype to proper datetime format
df['creationdate'] = pd.to_datetime(df.creationdate)

# Making sure that the column converted to datetime datatype
# print(df['creationdate'].dtype)

# 1. Find all questions that were created before 2014

df[df['creationdate'] < '2014-01-01 00:00:00']


# 2. Find all questions with a score more than 50

df[df['score'] > 50]

# 3. Find all questions with a score between 50 and 100

df[df['score'].between(50, 100)]

# 4. Find all questions answered by Scott Boston

# Making sure the case-insensivity

df[df['ans_name'].str.lower() == 'scott boston']

# 5. Find all questions answered by first 5 users

# Assigning first 5 unique persons in ans_name to answerers variable
answerers = df.dropna().drop_duplicates('ans_name').head()['ans_name']

# Returning all the questions that were answered by the 'answerers'
df[df['ans_name'].isin(answerers)]

# 6. Find all questions that were created between March, 2014 and October 2014 that were answered by Unutbu and have score less than 5.

df[
    (df['ans_name'].str.lower() == "unutbu") &
    (df['creationdate'].between('2014-03-01', '2014-10-31')) &
    (df['score'] < 5)
    ]

# 7. Find all questions that have score between 5 and 10 or have a view count of greater than 10,000

df[(df['score'].between(5, 10)) | (df['viewcount'] > 10_000)]

# 8. Find all questions that are not answered by Scott Boston

df[df['ans_name'].str.lower() != 'scott boston']
```

```python
import pandas as pd

df = pd.read_csv("titanic.csv")

# 1. Select Female Passengers in Class 1 with Ages between 20 and 30: Extract a DataFrame containing female passengers in Class 1 with ages between 20 and 30.

# Making sure that the "Sex" column contains only proper values
df['Sex'].value_counts()

# Filter the requirement with 'query()' method
min_age = 20
max_age = 30

df.query(
    'Sex == "female" and Pclass == 1 and @min_age <= Age <= @max_age'
)

# 2. Filter Passengers Who Paid More than $100: Create a DataFrame with passengers who paid a fare greater than $100.

new_df = df[df['Fare'] > 100]

new_df

# 3. Select Passengers Who Survived and Were Alone: Filter passengers who survived and were traveling alone (no siblings, spouses, parents, or children).

df[(df['Survived'] == 1) & (df['SibSp'] == 0) & (df['Parch'] == 0)]

# 4. Filter Passengers Embarked from 'C' and Paid More Than $50: Create a DataFrame with passengers who embarked from 'C' and paid more than $50.

embarked_from_C_paid_50 = df[(df['Embarked'] == "C") & (df['Fare'] > 50)]

embarked_from_C_paid_50

# 5. Select Passengers with Siblings or Spouses and Parents or Children: Extract passengers who had both siblings or spouses aboard and parents or children aboard.Select Passengers with Siblings or Spouses and Parents or Children: Extract passengers who had both siblings or spouses aboard and parents or children aboard.

df[(df['SibSp'] != 0) & (df['Parch'] != 0)]

# 6. Filter Passengers Aged 15 or Younger Who Didn't Survive: Create a DataFrame with passengers aged 15 or younger who did not survive.

unservived = df[(df['Age'] <= 15) & (df['Survived'] != 1)]

unservived

# 7. Select Passengers with Cabins and Fare Greater Than $200: Extract passengers with known cabin numbers and a fare greater than $200.

df[(df['Cabin'].notna()) & (df['Fare'] > 200)]

# 8. Filter Passengers with Odd-Numbered Passenger IDs: Create a DataFrame with passengers whose PassengerId is an odd number.

odd_passengerid = df[df['PassengerId'] % 2 != 0]

odd_passengerid

# 9. Select Passengers with Unique Ticket Numbers: Extract a DataFrame with passengers having unique ticket numbers.

unique_tickets = df[~df['Ticket'].duplicated(keep='first')]

unique_tickets

# 10. Filter Passengers with 'Miss' in Their Name and Were in Class 1: Create a DataFrame with female passengers having 'Miss' in their name and were in Class 1.

passengers_miss_class1 = df[(df['Name'].str.lower().str.contains('miss')) & (df['Pclass'] == 1)]

passengers_miss_class1


```
